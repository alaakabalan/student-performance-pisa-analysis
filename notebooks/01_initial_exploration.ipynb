```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df=pd.read_excel('../data/Maths.xlsx')

# First few rows
df.head()
```

```python
# Overview of the dataset
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# All columns
df.columns.tolist()
```

```python
# See data types
df.dtypes
```

```python
# 1. Distribution of final grades (G3)
sns.histplot(df['G3'], bins=15, kde=True)
plt.title('Distribution of final grades (G3)')
plt.show()
```

```python
# 2. Average G3 by Gender
sns.boxplot(data=df, x='sex', y='G3')
plt.show()
```

```python
# 3. Correlation heatmap
plt.figure(figsize=(14,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Do final grades differe with genders?
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='sex', y='G3')
plt.title(' Final grade by gender')
plt.xlabel('sex')
plt.ylabel('G3')
plt.show
```

```python
df.groupby("sex")['G3'].mean()
```

### Insight: Gender and Performance
On average, male students scored slightly higher than female students in their final Math grade.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='G3', hue='sex', kde=True, bins=15)
plt.title('Grade distribution by gender')
plt.xlabel('Final grade')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(data=df, x="sex", y="G3", palette="Pastel1")
plt.title("Distribution of Final Grades by Gender")
plt.xlabel("Gender")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
sns.countplot(data=df, x='G3', hue='sex')
plt.title("Count of Grades by Gender")
plt.show()
```

```python
# Does alcohol use affect the final grade?
df['avg_alcohol']=(df['Dalc']+df['Walc'])/2
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='avg_alcohol', y='G3', hue='sex')
plt.title('Final grade vs Average alcohol use')
plt.xlabel('Average alcohol use (1:low , 5:high)')
plt.ylabel('Final grade')
plt.show
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='avg_alcohol', y='G3')
plt.title('Final grade distr. by avg alcohol use')
plt.xlabel('Average alcohol use, 1:low 5:high')
plt.ylabel('G3')
plt.show()
```

### Insight: Does alcohol use affect performance and final grade?
The boxplot suggests that students who consume less alcohol tend to perform better. Higher levels of alcohol use are associated with a wider range of lower final grades, indicating a potential negative effect on academic achievement.

```python
import numpy as np
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Medu', y='G3', estimator=np.mean, errorbar='sd')
plt.title('Mother education vs Final grade')
plt.xlabel('Mother education')
plt.ylabel('Final grade G3')
plt.show()
```

```python
plt.Figure(figsize=(6, 4))
sns.barplot(data=df, x='Fedu', y='G3', estimator=np.mean, errorbar='sd')
plt.title('Fathers education vs Average final grade')
plt.xlabel('Father education')
plt.ylabel('Final grade')
plt.show()
```

```python
df['avg_parent_edu']=(df['Medu']+df['Fedu'])/2
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='avg_parent_edu', y='G3')
plt.title('Average parent education vs Final grade')
plt.xlabel('Average parents education 0=low 4=high')
plt.ylabel('Final grade G3')
plt.show()
```

```python
sns.boxplot(data=df, x='studytime', y='G3', hue='sex')
plt.title("Final grades by Study Time and Gender")
plt.show()
```

```python
sns.scatterplot(data=df, x='absences', y='G3', hue='sex')
plt.title('Final grade vs Number of absences')
plt.xlabel('Absences')
plt.ylabel('Final Grade')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='internet', y='G3')
plt.title("Final Grade by Internet Access")
plt.xlabel("Internet Access (yes/no)")
plt.ylabel("Final Grade (G3)")
plt.show()
```
