```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df=pd.read_excel('../data/Maths.xlsx')

# First few rows
df.head()
```

```python
# Overview of the dataset
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# All columns
df.columns.tolist()
```

```python
# See data types
df.dtypes
```

```python
# 1. Distribution of final grades (G3)
sns.histplot(df['G3'], bins=15, kde=True)
plt.title('Distribution of final grades (G3)')
plt.show()
```

```python
# 2. Average G3 by Gender
sns.boxplot(data=df, x='sex', y='G3')
plt.show()
```

```python
# 3. Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
